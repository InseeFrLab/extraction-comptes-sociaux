```python
from pathlib import Path

data_dir = "../src/data/marmot_data"
data = [path for path in (list(Path(data_dir).glob("*.png")) + list(Path(data_dir).glob("*.bmp")))]
data = [path for path in (list(Path(data_dir).glob("*.png")) + list(Path(data_dir).glob("*.bmp"))) if len(path.stem) < 13]
```

```python
data[:5]
```

```python
import numpy as np
from PIL import Image

image_path = data[8]
sample_id = image_path.stem
table_path = image_path.parent.parent.joinpath("table_mask", sample_id + ".bmp")
column_path = image_path.parent.parent.joinpath("column_mask", sample_id + ".bmp")

image = np.array(Image.open(image_path))
table_mask = np.expand_dims(np.array(Image.open(table_path)), axis=2)
column_mask = np.expand_dims(np.array(Image.open(column_path)), axis=2)
```

```python
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
plt.plot()
plt.imshow(image)
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.plot()
plt.imshow(image, interpolation='none')
plt.imshow(column_mask, interpolation='none', alpha=0.4)
plt.show()
```
