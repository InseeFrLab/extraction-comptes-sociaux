```python
import pandas as pd
```

```python
df = pd.read_csv("couverture_apet_acv.csv", sep=";")
```

```python
df
```

```python
regroupement_dict = {
    1: "Agences bancaires et immobilières",
    2: "Autres services",
    3: "Commerce d'équipement de la maison",
    4: "Commerce d'équipement de la personne",
    5: "Commerce de détail alimentaire",
    6: "Commerces et services automobiles",
    7: "Restauration et débits de boisson",
    8: "Services de coiffure et beauté"
}
```

```python
actif_dict = {
    0.0: "Pas actif selon l'INSEE",
    1.0: "Actif selon l'INSEE"
}
```

```python
df["apet_2018"] = df["apet_2018"].str[:2] + "." + df["apet_2018"].str[2:]
```

```python
df
```

```python
df.dtypes
```

```python
df["actif_et_2018"] = df["actif_et_2018"].map(actif_dict) 
```

```python
df
```

```python
df_regroupement = pd.read_csv("regroupement.csv", sep=";", header=None, names=["group_code", "apet"])
```

```python
df_regroupement
```

```python
df = df.merge(df_regroupement, left_on="apet_2018", right_on="apet")
```

```python
df
```

```python
df["group_label"] = df["group_code"].map(regroupement_dict) 
```

```python
grouped_df = df.groupby(["actif_et_2018", "group_label"])["nb_siret_present_cb", "nb_siret_present_insee"].sum()
```

```python
(100 * grouped_df["nb_siret_present_cb"] / grouped_df["nb_siret_present_insee"]).round(1)
```
